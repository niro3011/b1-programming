```python
from collections import Counter

login_attempts = [
    ("alice", "success"),
    ("bob", "failed"),
    ("bob", "failed"),
    ("charlie", "success"),
    ("bob", "failed"),
    ("alice", "failed")
]

# 1. Filtern: Nur fehlgeschlagene Benutzernamen extrahieren
failed_users = [user for user, result in login_attempts if result == "failed"]

# 2. Zählen: Die Häufigkeit jedes Benutzers zählen
failed_counts = Counter(failed_users)

print("Checking login attempts.....")

# 3. Prüfen: Auf 3 oder mehr fehlgeschlagene Versuche prüfen
for user, count in failed_counts.items():
    if count >= 3:
        print(f"ALERT: User '{user}' has {count} failed login attempts")

print("Security check complete")
```
